```python
def l2kg(lbs):
    return lbs * 0.453592


def kg2l(kg):
    return kg / 0.453592
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame()

df["kg"] = np.arange(100, 301, 10)
df["lbs"] = df["kg"].apply(kg2l)
df["lbs"] = df["lbs"].apply(lambda x: round(x, 2))

df
```
