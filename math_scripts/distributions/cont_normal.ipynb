### normal

The sum of many quantities from independent processes tend to distributed close to a normal distribution.

```python
from scipy.stats import norm
import matplotlib.pyplot as plt

# Define the distribution parameters
mean = 170
std_dev = 15

# Generate random samples from the distribution
samples = norm.rvs(size=1000, loc=mean, scale=std_dev)

# Calculate the PDF values
x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)
pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

# Plot the histogram of the samples
plt.hist(samples, bins=30, density=True, alpha=0.5, label='Samples')

# Plot the PDF
plt.plot(x, pdf, color='red', label='PDF')

# Add labels and legend
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.show()
```

### lognormal

Growth processes driven by the many small independent percentage changes tend to be lognormally distributed.

Exponential of a normal distribution is lognormally distributed.

```python
# Sample from the lognormal distribution
samples = np.random.lognormal(mean=0, sigma=1, size=1000)

# Calculate the PDF values
x = np.linspace(0, np.percentile(samples, 99.9), 100)
pdf = norm.pdf(np.log(x), loc=0, scale=1)

# Plot the histogram of the samples
plt.hist(samples, bins=30, density=True, alpha=0.5, label='Samples')

# Plot the PDF
plt.plot(x, pdf, color='red', label='PDF')

# Add labels and legend
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.show()
```

### logistic distribution

A random variable with a logistic function as its CDF is logistic distributed.

The error of the log odds ratio of a logistic regression is logistic distributed.

```python
from scipy.stats import logistic
import matplotlib.pyplot as plt
import numpy as np

# Define the location and scale parameters
loc = 0
scale = 1

# Generate random samples from the logistic distribution
samples = logistic.rvs(loc=loc, scale=scale, size=1000)

# Calculate the PDF values
x = np.linspace(-10, 10, 100)
pdf = logistic.pdf(x, loc=loc, scale=scale)

# Plot the histogram of the samples
plt.hist(samples, bins=30, density=True, alpha=0.5, label='Samples')

# Plot the PDF
plt.plot(x, pdf, color='red', label='PDF')
# compare with normal distribution
pdf_normal = norm.pdf(x, loc=loc, scale=scale)
plt.plot(x, pdf_normal, color='green', label='PDF normal')

# Add labels and legend
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.show()
```

### chi squared

The sum of squares of many independent normal random variables is chi squared distributed.

```python
from scipy.stats import chi2
import matplotlib.pyplot as plt

# Define the degrees of freedom
df = 5

# Generate random samples from the chi-squared distribution
samples = chi2.rvs(df=df, size=1000)

# Calculate the PDF values
x = np.linspace(0, 20, 100)
pdf = chi2.pdf(x, df=df)

# Plot the histogram of the samples
plt.hist(samples, bins=30, density=True, alpha=0.5, label='Samples')

# Plot the PDF
plt.plot(x, pdf, color='red', label='PDF')

# Add labels and legend
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.show()
```

### t

A distribution that is similiar to normal distribution, but with heavier tails and a lower peak, could be modeled by the T distribution.

The ratio of a normal random variable and the square root of a chi squared random variable is t distributed.

```python
from scipy.stats import t
import matplotlib.pyplot as plt
import numpy as np

# Define the degrees of freedom
df = 10

# Generate random samples from the T distribution
samples = t.rvs(df=df, size=1000)

# Calculate the PDF values
x = np.linspace(-4, 4, 100)
pdf = t.pdf(x, df=df)

# Plot the histogram of the samples
plt.hist(samples, bins=30, density=True, alpha=0.5, label='Samples')

# Plot the PDF
plt.plot(x, pdf, color='red', label='PDF')
# add comparison with normal distribution
pdf_norm = norm.pdf(x, loc=0, scale=1)
plt.plot(x, pdf_norm, color='green', label='Normal PDF')

# Add labels and legend
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.show()
```

### F

The ratio of two chi squared random variables is F distributed.

```python
from scipy.stats import f
import matplotlib.pyplot as plt
import numpy as np

# Define the degrees of freedom
dfn = 5  # numerator degrees of freedom
dfd = 10  # denominator degrees of freedom

# Generate random samples from the F distribution
samples = f.rvs(dfn=dfn, dfd=dfd, size=1000)

# Calculate the PDF values
x = np.linspace(0, 15, 100)
pdf = f.pdf(x, dfn=dfn, dfd=dfd)

# Plot the histogram of the samples
plt.hist(samples, bins=30, density=True, alpha=0.5, label='Samples')

# Plot the PDF
plt.plot(x, pdf, color='red', label='PDF')
# the chi-squared distributions
pdf_chi2 = chi2.pdf(x, df=dfn)
plt.plot(x, pdf_chi2, color='green', label='Chi-squared PDF (numerator)')
pdf_chi2_2 = chi2.pdf(x, df=dfd)
plt.plot(x, pdf_chi2_2, color='blue', label='Chi-squared PDF (denominator)')

# Add labels and legend
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.show()
```

### pareto

The income or wealth of a population is often modeled by a Pareto distribution.
These reflects a large proportion of wealth are held by a small proportion of the population.

```python
from scipy.stats import pareto
import matplotlib.pyplot as plt
import numpy as np

# Define the shape parameter
alpha = 2

# Generate random samples from the Pareto distribution
samples = pareto.rvs(b=alpha, size=1000)

# Calculate the PDF values
x = np.linspace(1, 20, 100)
pdf = pareto.pdf(x, b=alpha)

# Plot the histogram of the samples
plt.hist(samples, bins=30, density=True, alpha=0.5, label='Samples')

# Plot the PDF
plt.plot(x, pdf, color='red', label='PDF')

# Add labels and legend
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.show()
```

### burr

Generalization of Pareto distribution and log-logistic distribution.
Could be used to model household income.

```python
from scipy.stats import burr
import matplotlib.pyplot as plt
import numpy as np

# Define the shape parameters
c = 2
k = 5

# Generate random samples from the Burr distribution
samples = burr.rvs(c=c, d=k, size=1000)

# Calculate the PDF values
x = np.linspace(0, 10, 100)
pdf = burr.pdf(x, c=c, d=k)

# Plot the histogram of the samples
plt.hist(samples, bins=30, density=True, alpha=0.5, label='Samples')

# Plot the PDF
plt.plot(x, pdf, color='red', label='PDF')

# Add labels and legend
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.show()
```
