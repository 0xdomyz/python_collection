```python
import matplotlib.pyplot as plt
from scipy.stats import t
from scipy.stats import norm
import numpy as np
```

```python
df = 100
x = np.linspace(t.ppf(0.001, df), t.ppf(0.999, df), 1000)
plt.plot(x, t.pdf(x, df), "r-", label="t pdf")
plt.plot(x, norm.pdf(x), "b-", label="norm pdf")
plt.title(f"T-distribution with {df} degrees of freedom")
plt.show()
```
