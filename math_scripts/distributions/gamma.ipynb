```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.api as sm
from scipy import stats
```

### alpha beta defn

```python
alpha = 3
beta = 10

def examine(alpha, beta):
    rv = stats.gamma

    x = np.linspace(0, 1, 100)
    p = rv.pdf(x, alpha, 0, 1/beta)
    c = rv.cdf(x, alpha, 0, 1/beta)

    print(f"{rv.stats(alpha, 0, 1/beta, moments = 'm') = }")
    print(f"{rv.stats(alpha, 0, 1/beta, moments = 'v') = }")
    print(f"{rv.stats(alpha, 0, 1/beta, moments = 's') = }")
    print(f"{rv.stats(alpha, 0, 1/beta, moments = 'k') = }")

    plt.plot(x, p, label='pdf', color='red')
    plt.plot(x, c, label='cdf', color='blue')
    plt.legend()

examine(alpha = 3, beta = 10)
```

```python
examine(alpha = 5, beta = 10)
```

```python
examine(alpha = 5, beta = 100)
```

```python
examine(alpha = 15, beta = 10)
```
