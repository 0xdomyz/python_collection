```python
import numpy as np
import matplotlib.pyplot as plt
```

### exponential

Distance between events in a Poisson process is exponentially distributed.

```python
from scipy.stats import expon

# parameters
lam = 0.5

samples = expon.rvs(scale=1 / lam, size=1000)
plt.hist(samples, bins=100, density=True, alpha=0.5, label="Simulation")

# PDF
x = np.linspace(0, 8, 100)
pdf = expon.pdf(x, scale=1 / lam)

plt.plot(x, pdf, "ro-", label="Analytical PDF")
plt.xlabel("x")
plt.ylabel("Probability")
plt.title("Exponential Distribution")
plt.legend()
plt.show()
```

### weibull

Time to failure of a component is Weibull distributed. The failure rate could be constant, increasing or decreasing over time.

```python
from scipy.stats import weibull_min

# parameters
c = 1.5
loc = 0
scale = 2

samples = weibull_min.rvs(c, loc, scale, size=1000)
plt.hist(samples, bins=100, density=True, alpha=0.5, label="Simulation")

# PDF
x = np.linspace(0, 8, 100)
pdf = weibull_min.pdf(x, c, loc, scale)

plt.plot(x, pdf, "ro-", label="Analytical PDF")
plt.xlabel("x")
plt.ylabel("Probability")
plt.title("Weibull Distribution")
plt.legend()
plt.show()
```

### beta

A random variable that is bounded between 0 and 1 can have a beta distribution. It is often used to model probabilities.

```python
from scipy.stats import beta

# parameters
a = 2
b = 5
loc = 0

samples = beta.rvs(a, b, loc, size=1000)
plt.hist(samples, bins=100, density=True, alpha=0.5, label="Simulation")

# PDF
x = np.linspace(0, 1, 100)
pdf = beta.pdf(x, a, b, loc)

plt.plot(x, pdf, "ro-", label="Analytical PDF")
plt.xlabel("x")
plt.ylabel("Probability")
plt.title("Beta Distribution")
plt.legend()
plt.show()
```

### uniform

An experiment that is equally probable to result within a given range can be modeled with a uniform distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
low = 1
high = 10
size = 1000

# Generate random samples from a uniform distribution
samples = np.random.uniform(low, high, size)

# Plot histogram of the samples
plt.hist(samples, bins=100, density=True, alpha=0.5, label="Simulation")

# Plot the PDF
x = np.linspace(low, high, 100)
pdf = np.ones_like(x) / (high - low)
plt.plot(x, pdf, "ro-", label="Analytical PDF")

# Set plot labels and title
plt.xlabel("x")
plt.ylabel("Probability")
plt.title("Uniform Distribution")

# Show legend and plot
plt.legend()
plt.show()
```

### uniform 0-1

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random samples from a uniform distribution
samples = np.random.uniform(0, 1, 1000)

# Plot histogram of the samples
plt.hist(samples, bins=100, density=True, alpha=0.5, label="Simulation")

# Plot the PDF
x = np.linspace(0, 1, 100)
pdf = np.ones_like(x)
plt.plot(x, pdf, "ro-", label="Analytical PDF")

# Set plot labels and title
plt.xlabel("x")
plt.ylabel("Probability")
plt.title("Uniform Distribution (0 to 1)")

# Show legend and plot
plt.legend()
plt.show()
```

### discrete uniform

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
low = 1
high = 10
size = 1000

# Generate random samples from a discrete uniform distribution
samples = np.random.choice(np.arange(low, high + 1), size)

# Compute the PMF
pmf = np.ones(high - low + 1) / (high - low + 1)

# Plot histogram of the samples
plt.hist(
    samples, bins=np.arange(low, high + 2), density=True, alpha=0.5, label="Simulation"
)

# Plot the PMF
plt.plot(np.arange(low, high + 1), pmf, "ro-", label="PMF")

# Set plot labels and title
plt.xlabel("x")
plt.ylabel("Probability")
plt.title("Discrete Uniform Distribution")

# Show legend and plot
plt.legend()
plt.show()
```
