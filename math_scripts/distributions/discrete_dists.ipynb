```python
import numpy as np
import matplotlib.pyplot as plt
```

### geometric

Given a series of iid bernoulli trials, the number of trails conducted till the first success, is of the geometric distribution.

```python
from scipy.stats import geom

# parameters
p = 0.3  # Probability of success 

samples = geom.rvs(p, size=1000)
plt.hist(samples, bins=range(1, np.max(samples)+2), density=True, alpha=0.5, label='Simulation')

# PMF
x = np.arange(1, np.max(samples)+1)
pmf = geom.pmf(x, p)

plt.plot(x, pmf, 'ro-', label='Analytical PMF')
plt.xlabel('Number of trials until success')
plt.xticks(np.arange(1, np.max(samples)+1, 1.0))
plt.ylabel('Probability')
plt.title('Geometric Distribution')
plt.legend()
plt.show()
```

### negative binomial

Given series of iid bernoulli trials, the number of trails conducted till the rth success, is of the negative binomial distribution.

```python
from scipy.stats import nbinom

# parameters
n = 3  # Number of successes
p = 0.3  # Probability of success

samples = nbinom.rvs(n, p, size=1000)
plt.hist(samples, bins=range(0, np.max(samples)+2), density=True, alpha=0.5, label='Simulation')

# PMF
x = np.arange(0, np.max(samples)+1)
pmf = nbinom.pmf(x, n, p)

plt.plot(x, pmf, 'ro-', label='Analytical PMF')
plt.xlabel('Number of trials until n successes')
plt.ylabel('Probability')
plt.title('Negative Binomial Distribution')
plt.legend()
plt.show()
```

### poisson

The number of events occurring in a fixed interval of time or space, given the occurance is of constant average rate, and independent of time from last event, is of the poisson distribution.

```python
from scipy.stats import poisson

# parameters
mu = 3  # mean

samples = poisson.rvs(mu, size=1000)
plt.hist(samples, bins=range(0, np.max(samples)+2), density=True, alpha=0.5, label='Simulation')

# PMF
x = np.arange(0, np.max(samples)+1)
pmf = poisson.pmf(x, mu)

plt.plot(x, pmf, 'ro-', label='Analytical PMF')
plt.xlabel('Number of events')
plt.ylabel('Probability')
plt.title('Poisson Distribution')
plt.legend()
plt.show()
```

### binomial

Given series of iid bernoulli trials, the number of successes in n trials, is of the binomial distribution.

```python
from scipy.stats import binom
import matplotlib.pyplot as plt

# Parameters
n = 10  # Number of trials
p = 0.5  # Probability of success

# Generate a sample
sample = binom.rvs(n, p, size=1000)

# Plot the histogram
plt.hist(sample, bins=range(0, n+2), density=True, alpha=0.5, label='Simulation')

# Plot the PMF
x = range(0, n+1)
pmf = binom.pmf(x, n, p)
plt.plot(x, pmf, 'ro-', label='Analytical PMF')

# Set labels and title
plt.xlabel('Number of successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution')

# Show the legend and plot
plt.legend()
plt.show()

```

### bernoulli

If a random variable X takes value 1 with probability p and value 0 with probability q=1-p, then X is said to follow bernoulli distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
p = 0.3  # Probability of success
size = 1000  # Sample size

# Generate a sample
sample = np.random.choice([0, 1], size=size, p=[1-p, p])

# Plot the histogram
plt.hist(sample, bins=[0, 1, 2], density=True, alpha=0.5, label='Simulation')

# Set labels and title
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Bernoulli Distribution')

# Show the legend and plot
plt.legend()
plt.show()
```
