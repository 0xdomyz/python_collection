### moments

#### The kth raw moment:

\begin{align*}
\mu_k &= E(X^k) \\
      &= \frac{1}{n} \sum_{i=1}^{n} x_i^k
\end{align*}

#### The kth central moment:

\begin{align*}
m_k &= E((X - \mu)^k) \\
      &= \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^k
\end{align*}

#### The kth standardized moment:

\begin{align*}
\bar{m_k} &= E\left(\left(\frac{X - \mu}{\sigma}\right)^k\right) \\
      &= \frac{1}{n} \sum_{i=1}^{n} \left(\frac{x_i - \bar{x}}{s}\right)^k
\end{align*}

### mean, variance, skewness, kurtosis

#### mean

mean is the first raw moment:

$$\mu_1 = E(X)$$

#### variance

variance is the second central moment:

$$\sigma^2 = m_2 = E((X - \mu)^2)$$

Also variance is the second raw moment minus the square of the mean:

$$\sigma^2 = \mu_2 - \mu_1^2$$


#### skewness

skewness is the third standardized moment:

$$\bar{m_3} = E\left(\left(\frac{X - \mu}{\sigma}\right)^3\right)$$

It measures the relative size of the two tails, since values more than one standard deviation from the mean contribute to the skewness excess of 1.

#### kurtosis

kurtosis is the fourth standardized moment:

$$\bar{m_4} = E\left(\left(\frac{X - \mu}{\sigma}\right)^4\right)$$

it measures the relative size of the tails, since values more than one standard deviation from the mean contribute to the kurtosis more.

excess kurtosis is the kurtosis minus 3.

kurtois is also the variance of the squared z-scores plus 1:

$$\bar{m_4} = Var\left(Z^2\right) + 1 $$

where

$$Z = \frac{X - \mu}{\sigma}$$



### exmaples

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Parameters of the gamma distribution
shape = 2  # shape parameter
scale = 2  # scale parameter

# Generate x values
x = np.linspace(0, 10, 100)

# Calculate the PDF values
pdf = gamma.pdf(x, shape, scale)

# Plot the PDF
plt.plot(x, pdf)
plt.xlabel("x")
plt.ylabel("PDF")
plt.title("Theoretical PDF of Gamma Distribution")
plt.grid(True)
plt.show()
```

```python
# simulate gamma distribution
n = 10000
s = gamma.rvs(shape, scale=scale, size=n)

# Plot the histogram
count, bins, ignored = plt.hist(s, 50, density=True)
plt.plot(x, pdf, linewidth=2, color="r")
plt.xlabel("x")
plt.ylabel("PDF")
plt.title("Histogram of Gamma Distribution")
plt.grid(True)
plt.show()
```
