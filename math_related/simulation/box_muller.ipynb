```python
import numpy as np
import matplotlib.pyplot as plt

def box_muller(n_samples=10000):
    # Generate uniform random numbers
    u1 = np.random.uniform(0, 1, n_samples)
    u2 = np.random.uniform(0, 1, n_samples)
    
    # Box-Muller transformation
    z1 = np.sqrt(-2 * np.log(u1)) * np.cos(2 * np.pi * u2)
    z2 = np.sqrt(-2 * np.log(u1)) * np.sin(2 * np.pi * u2)
    
    return z1, z2

# Generate samples
z1, z2 = box_muller(10000)
```

```python

# Plot to verify normality
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Histograms
axes[0,0].hist(z1, bins=50, alpha=0.7, density=True, label='Box-Muller Z1')
axes[0,0].plot(np.linspace(-4, 4, 100), 
               (1/np.sqrt(2*np.pi)) * np.exp(-0.5 * np.linspace(-4, 4, 100)**2), 
               'r-', label='N(0,1)')
axes[0,0].set_title('Box-Muller Z1')
axes[0,0].legend()

axes[0,1].hist(z2, bins=50, alpha=0.7, density=True, label='Box-Muller Z2')
axes[0,1].plot(np.linspace(-4, 4, 100), 
               (1/np.sqrt(2*np.pi)) * np.exp(-0.5 * np.linspace(-4, 4, 100)**2), 
               'r-', label='N(0,1)')
axes[0,1].set_title('Box-Muller Z2')
axes[0,1].legend()


# Q-Q plots
from scipy import stats
stats.probplot(z1, dist="norm", plot=axes[1,0])
axes[1,0].set_title('Q-Q Plot Z1')

stats.probplot(z2, dist="norm", plot=axes[1,1])
axes[1,1].set_title('Q-Q Plot Z2')

plt.tight_layout()
```

```python

print(f"Z1: Mean={np.mean(z1):.3f}, Std={np.std(z1):.3f}")
print(f"Z2: Mean={np.mean(z2):.3f}, Std={np.std(z2):.3f}")
```
