```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.api as sm
from scipy import stats
```

### alpha beta defn

```python
alpha = 3
beta = 10
```

```python
def examine(alpha, beta):
    rv = stats.gamma

    large_pctil = rv.ppf(0.995, alpha, 0, 1 / beta)

    x = np.linspace(0, large_pctil, 1000)
    p = rv.pdf(x, alpha, 0, 1 / beta)
    c = rv.cdf(x, alpha, 0, 1 / beta)

    print(f"{rv.stats(alpha, 0, 1/beta, moments = 'mv') = }")
    print(f"{rv.stats(alpha, 0, 1/beta, moments = 'sk') = }")
    print(f"{rv.ppf(0.01,alpha, 0, 1/beta) = }")
    print(f"{rv.ppf(0.99,alpha, 0, 1/beta) = }")

    # plot
    fig, ax1 = plt.subplots()

    ax1.plot(x, p, label="pdf", color="red")
    ax1.set_xlabel("X")
    ax1.set_ylabel("PDF")

    ax2 = plt.gca().twinx()
    ax2.plot(x, c, label="cdf", color="blue")
    ax2.set_ylabel("CDF")

    fig.legend(loc="upper left")

    plt.show()


examine(alpha=20, beta=1 / 10e3)
```
