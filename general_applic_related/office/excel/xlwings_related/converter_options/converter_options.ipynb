```python
import numpy as np
import pandas as pd
import seaborn as sns
import xlwings as xw

df = sns.load_dataset('titanic')
print(f"{df.shape = }")
print(df.head().to_string())
```

```python
import xlwings as xw

wb = xw.Book()
ws = wb.sheets.add('df')
ws["A1"].value = df
ws.tables.add(source=ws["A1"].expand())
```

 ## Converter Options
 ####################################################################################################

```python
lst = ws["A1:C3"].options().value
lst
```

```python
df2 = ws["A1:C3"].options(pd.DataFrame).value
df2
```

```python
d = ws["A1:B3"].options(dict).value
d
```

```python
t = ws["A1:B3"].options(tuple).value
t
```

```python
a = ws["A1:B3"].options(np.array).value
a
```

 ## ndim
 ####################################################################################################

```python
ws["A1"].options().value # scalar
```

```python
ws["A1"].options(ndim=2).value # 2D array
```

```python
ws["A1:A3"].options(list).value # 1D array
```

```python
ws["A1:A3"].options(list,ndim=2).value # 2D array, not 1D array
```
