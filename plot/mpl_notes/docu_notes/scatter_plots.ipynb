```python
import numpy as np
import matplotlib.pyplot as plt
```

### via plot

```python
# plot via specifiying marker
x = np.linspace(0, 10, 30)
y = np.sin(x)
plt.plot(x, y, "o", color="black");
```

```python
rng = np.random.RandomState(0)
for i, marker in enumerate(["o", ".", ",", "x", "+", "v", "^", "<", ">", "s", "d"]):
    plt.plot(i, i, marker, label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
```

```python
# can be mixed with lines
plt.plot(x, y, "-ok");
```

```python
# other marker configs
plt.plot(
    x,
    y,
    "-p",
    color="gray",
    markersize=15,
    linewidth=4,
    markerfacecolor="white",
    markeredgecolor="gray",
    markeredgewidth=2,
)
plt.ylim(-1.2, 1.2);
```

### via scatter

```python
plt.scatter(x, y, marker="o");
```

```python
# control each points individually
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
plt.scatter(
    x,
    y,
    c=colors,
    s=sizes,
    alpha=0.3,
    # cmap='viridis'
)
plt.colorbar();  # show color scale
```

```python
# visualise more dimensions
from sklearn.datasets import load_iris

iris = load_iris()
features = iris.data.T
plt.scatter(
    features[0],
    features[1],
    alpha=0.2,
    s=100 * features[3],
    c=iris.target,
    cmap="viridis",
)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```
