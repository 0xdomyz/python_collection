```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.api as sm
from scipy import stats
```

```python
import seaborn as sns
```

## Line charts

```python
# a line
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
```

```python
# pylab interface
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x))
```

```python
# multiple objects
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
```

### customisation

```python
# color
x = np.linspace(0, 10, 1000)

plt.plot(x, np.sin(x - 0), color="blue")  # specify color by name
plt.plot(x, np.sin(x - 1), color="g")  # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color="0.75")  # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color="#FFDD44")  # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3))  # RGB tuple, values 0 and 1
plt.plot(x, np.sin(x - 5), color="chartreuse");  # all HTML color names supported
```

```python
x = np.linspace(0, 10, 1000)

plt.plot(x, 0.2 * x + 0, color="r")
plt.plot(x, 0.2 * x + 1, color="g")
plt.plot(x, 0.2 * x + 2, color="b")
plt.plot(x, 0.2 * x + 3, color="c")
plt.plot(x, 0.2 * x + 4, color="m")
plt.plot(x, 0.2 * x + 5, color="y")
plt.plot(x, 0.2 * x + 6, color="k")
```

```python
# style
x = np.linspace(0, 10, 1000)

plt.plot(x, x + 0, linestyle="solid")
plt.plot(x, x + 1, linestyle="dashed")
plt.plot(x, x + 2, linestyle="dashdot")
plt.plot(x, x + 3, linestyle="dotted")
# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle="-")  # solid
plt.plot(x, x + 5, linestyle="--")  # dashed
plt.plot(x, x + 6, linestyle="-.")  # dashdot
plt.plot(x, x + 7, linestyle=":")  # dotted
```

```python
# color and style as nono kw arguments
x = np.linspace(0, 10, 1000)
plt.plot(x, x + 0, "-g")  # solid green
plt.plot(x, x + 1, "--c")  # dashed cyan
plt.plot(x, x + 2, "-.k")  # dashdot black
plt.plot(x, x + 3, ":r");  # dotted red
```

```python
# axis limits
x = np.linspace(0, 10, 1000)

plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

```python
# reverse limits
plt.plot(x, np.sin(x))
plt.xlim(10, 0)
plt.ylim(1.2, -1.2)
```

```python
# in single call
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5])
```

```python
# higher lvel control via axis
plt.plot(x, np.sin(x))
plt.axis("tight");
```

```python
plt.plot(x, np.sin(x))
plt.axis("equal");
```

```python
# title and labels
x = np.linspace(0, 10, 1000)

plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

```python
# multiple lines with title and labels
x = np.linspace(0, 10, 1000)

plt.plot(x, np.sin(x), "-g", label="sin(x)")
plt.plot(x, np.cos(x), ":b", label="cos(x)")
plt.axis("equal")

plt.legend()
```

```python
# object oriented interface
fig, ax = plt.subplots(1)

x = np.linspace(0, 10, 1000)

ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))

ax.set_title("Sine and Cosine Curves")
ax.set_xlabel("x")
ax.set_ylabel("sin(x)")
ax.set_xlim(-1, 11)
ax.set_ylim(-1.5, 1.5)
ax.legend(["sin(x)", "cos(x)"])
```

```python
# object oriented interface via set method
fig, ax = plt.subplots(1)

x = np.linspace(0, 10, 1000)

ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))

ax.set(
    title="Sine and Cosine Curves",
    xlabel="x",
    ylabel="sin(x)",
    xlim=(-1, 11),
    ylim=(-1.5, 1.5),
)
ax.legend(["sin(x)", "cos(x)"])
```
