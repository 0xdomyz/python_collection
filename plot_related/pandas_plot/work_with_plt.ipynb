```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 1000)
y = np.sin(x)

df = pd.DataFrame({"x": x, "y": y})
print(f"{df.shape}")
df.head(1)
```

```python
# extrat the fig, ax used by pd.DataFrame.plot
fig, ax = plt.subplots()
df.plot(kind="line", x="x", y="y", ax=ax)
ax.set_title("y = sin(x)")
plt.show()
```
