## What a tree is
A tree is a hierarchical, acyclic structure with:
- a root
- parent-child relationships
- subtrees
- depth and height

## Why trees exist
Trees encode hierarchy, sorted order, and search structure.
They solve: “I need fast lookup, insertion, and deletion while preserving structure.”

## Operational behaviour
- Traversals:
  - preorder
  - inorder
  - postorder
  - level-order (BFS)
- Performance depends on height:
  - balanced → O(log n)
  - unbalanced → O(n)

## Important tree types
- Binary Search Tree (BST)
- AVL / Red-Black Trees
- Heaps
- Tries
- B-Trees (databases)
- Segment trees

## Where trees appear
- Database indexes
- DOM structure
- Routing tables
- Compiler ASTs
- Filesystems
- UI component hierarchies


```python
# Minimal binary tree node
class Node:
    def __init__(self, v):
        self.v = v
        self.left = None
        self.right = None
```

```python
# Insert into a BST
def insert(root, x):
    if root is None:
        return Node(x)
    if x < root.v:
        root.left = insert(root.left, x)
    else:
        root.right = insert(root.right, x)
    return root
```

```python
# Inorder traversal (sorted output)
def inorder(root):
    if root:
        yield from inorder(root.left)
        yield root.v
        yield from inorder(root.right)
```

```python
# Trie (prefix tree) insert
class TrieNode:
    def __init__(self):
        self.children = {}
        self.end = False

def insert(root, word):
    node = root
    for ch in word:
        node = node.children.setdefault(ch, TrieNode())
    node.end = True
```
