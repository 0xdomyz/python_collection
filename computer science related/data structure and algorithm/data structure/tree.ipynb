## What a tree is
A tree is a hierarchical, acyclic structure with:
- a root
- parent-child relationships
- one parent except the root

## Why trees exist
Trees encode hierarchy, sorted order, and search structure.
They solve: “I need fast lookup, insertion, and deletion while preserving structure.”

## Operational behaviour
- Traversals:
  - preorder
  - inorder
  - postorder
  - level-order (BFS)
- Performance depends on height:
  - balanced → O(log n)
  - unbalanced → O(n)

## Important tree types
- Binary Search Tree (BST)
- AVL / Red-Black Trees
- Heaps
- Tries
- B-Trees (databases)
- Segment trees

## Where trees appear
- Database indexes
- DOM structure
- Routing tables
- Compiler ASTs
- Filesystems
- UI component hierarchies

````python
# Minimal binary tree node
class Node:
    def __init__(self, v):
        self.v = v
        self.left = None
        self.right = None

# Insert into a BST
def insert(root, x):
    if root is None:
        return Node(x)
    if x < root.v:
        root.left = insert(root.left, x)
    else:
        root.right = insert(root.right, x)
    return root

# Inorder traversal (sorted output)
def inorder(root):
    if root:
        yield from inorder(root.left)
        yield root.v
        yield from inorder(root.right)
````

````python
root = Node(5)
insert(root, 3)
insert(root, 7)
insert(root, 4)

for i in inorder(root): 
    print(i)
````

````python
# Trie (prefix tree) insert
class TrieNode:
    def __init__(self):
        self.children = {}
        self.end = False

def insert(root, word):
    node = root
    for ch in word:
        node = node.children.setdefault(ch, TrieNode())
    node.end = True

def search(root, word):
    node = root
    for ch in word:
        if ch not in node.children:
            return False
        node = node.children[ch]
    return node.end
````

````python
root = TrieNode()
insert(root, "cat")
insert(root, "car")
insert(root, "cart")

print(search(root, "car"))  # True
````

````python
from collections import deque

def level_order(root):
    if root is None: return []
    q = deque([root]); out = []
    while q:
        node = q.popleft()
        out.append(node.v)
        if node.left: q.append(node.left)
        if node.right: q.append(node.right)
    return out

level_order(root)
````

````python
def height(node):
    if node is None: return 0
    return 1 + max(height(node.left), height(node.right))

def is_balanced(node):
    if node is None: return True
    lh, rh = height(node.left), height(node.right)
    return abs(lh - rh) <= 1 and is_balanced(node.left) and is_balanced(node.right)

height(root), is_balanced(root)
````
