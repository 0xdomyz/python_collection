## Control Flow and Functions
- Conditionals and loops shape program flow.
- Small, pure functions are easiest to test and reuse.

```python
def fizzbuzz(n: int) -> list[str]:
    out = []
    for i in range(1, n + 1):
        if i % 15 == 0: out.append("fizzbuzz")
        elif i % 3 == 0: out.append("fizz")
        elif i % 5 == 0: out.append("buzz")
        else: out.append(str(i))
    return out

fizzbuzz(15)
```

```python
from dataclasses import dataclass
import math

@dataclass
class Point:
    x: float
    y: float

    def mag(self) -> float:
        return math.hypot(self.x, self.y)

p = Point(3, 4)
p.mag()  # 5.0
```

## Testing and Debugging Notes
- Prefer small pure functions; assert on inputs/outputs.
- Use `try/except` narrowly; log and re-raise for unexpected cases.
- Write a quick self-check harness inside the notebook before formal tests.

## Modularity and State
- Prefer functions that take data in and return data out; minimize hidden state.
- Group related behavior via small classes or modules; avoid “god objects.”
- Add type hints and docstrings to document contracts and enable tooling.

```python
def add_tax(amount: float, rate: float = 0.1) -> float:
    """Pure function: returns amount with tax."""
    return round(amount * (1 + rate), 2)

# Quick self-checks
assert add_tax(100, 0.05) == 105.0
assert add_tax(19.99, 0.1) == 21.99
add_tax(42, 0.2)
```
