```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
capital = 5
prob = 0.51
n_sim = 10_000

t = np.arange(0, 10000, 1)
bins = np.random.binomial(1, prob, size=(len(t), n_sim))
v = capital + np.cumsum(2 * bins - 1, axis=0)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 6))

ax.plot(t, v[:, 0:10], color='C0', alpha=0.7)
ax.plot(t, 0 * t, color='k', lw=2)
```

```python
is_ruin = (v <= 0).any(axis=0)
prob_ruin = is_ruin.sum() / n_sim
prob_ruin
```

```python
prob_ruin_limit = ((1 - prob) / prob) ** capital
prob_ruin_limit
```
