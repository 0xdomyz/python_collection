### does awaiting coro cede control? No

```python
import asyncio


async def coro1():
    print("coro1: start")
    await asyncio.sleep(1)
    print("coro1: end")


async def coro2():
    print("coro2: start")
    await asyncio.sleep(0.5)
    print("coro2: end")


async def main():
    # Awaiting coroutines directly, not wrapping in tasks
    await coro1()
    await coro2()


async def concurrent_main():
    # Awaiting both coroutines concurrently
    await asyncio.gather(coro1(), coro2())


print("=== Sequential (no concurrency) ===")
await main()

print("\n=== Concurrent (event loop switches between coros) ===")
await concurrent_main()
```
