### asyncio goals

- maximize a single thread's usage
- by asynchronously handle I/O
- by using coroutines (which can be ran in small chunks), instead of blocking I/O

### The bartender analogy

- A bartender (event loop) is responsible for serving multiple customers (coroutines/tasks) efficiently.
- Even though there is only 1 bartender, they can handle multiple customers by switching between them based on signals (await).
- This is due to the fact that customers take time to do things on their own (like drinking or thinking), during which the bartender may be able to attend to other customers.
Concepts Comparison

| Bartender Analogy Concepts |  Asyncio Concept
|-----------------|-------------------
| Bartender, common sense, Bartending SOP | Event Loop, asyncio, python
| customer | coroutine / task
| customer doing something but no signal for bartender to walk away yet | await for coroutine
| customer signalling bartender to walk away | await for task
