### Tasks

- Coro tied to event loop
    - It maintains list of callback functions (A), which are used in the event loop
        - A callback is a subroutine function which is passed as an argument to be executed at some point in the future.
    - On creation (B), it is auto scheduled for execution, by adding a callback to run it in the event loop's to-do list
        - Note it is not run immediately and need to wait for event loop to pick it up
        - Note the task itself is not added to the event loop, so there is slight chance of having gc related issues
- gathered in a collection to have strong ref for avoiding gc

```python
import asyncio
```

```python
async def aprint(magic_number: int):
    print(f"The magic number is: {magic_number}")

coroutine = aprint(magic_number=5)

task = asyncio.create_task(coroutine) # B

# async def main():
#     await task

# asyncio.run(main()) # manage event loop and ensure coro finishes
```
