### event loop

- It contains a 'collection of jobs' / 'to-do list' to be run
- It works like a loop:
    - Invoke/Select/give control/call/run a job
    - When the job pauses/completes, control is returned to the event loop
    - Repeat endlessly
- If no job is pending execution, the loop is smart enough to simply rest and resume when needed.
- Note efficiency is affected by how well jobs share controls, i.e. not hogging the control.

```python
import asyncio

# This creates an event loop and indefinitely cycles through
# its collection of jobs.
event_loop = asyncio.new_event_loop()
# event_loop.run_forever()
```
